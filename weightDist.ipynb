```python
dataXlsxPath = "tool_hook_fish_20250528172514.xlsx"
sheetName = "tool_hook_fish_20250528172514"

import pandas as pd

df = pd.read_excel(dataXlsxPath, sheet_name=sheetName)

df.head()
```

```python
# 只分析 “鱼名称” 列 为 “Bigmouth_Buffalo_Trophy” 的行
# 计算 “Bigmouth_Buffalo_Trophy” 出现的次数
dfBuffaloTrophy = df[df['鱼名称'] == 'Bigmouth_Buffalo_Trophy']
countBuffaloTrophy = len(dfBuffaloTrophy)
print("Bigmouth_Buffalo_Trophy 出现的次数为：", countBuffaloTrophy)
```

```python
# 画出其中的 重量(kg) 列 和 长度(cm) 列的二维关联曲线，横轴为长度，纵轴为重量
import matplotlib.pyplot as plt
import numpy as np

# 使用dfBuffaloTrophy中的数据
# dfBuffaloTrophy = pd.read_csv('buffaloTrophy.csv')
# dfBuffaloTrophy = dfBuffaloTrophy.dropna()
# dfBuffaloTrophy = dfBuffaloTrophy.reset_index(drop=True)
# 画出其中的 重量(kg) 列 和 长度(cm) 列的二维关联曲线，横轴为长度，纵轴为重量
plt.scatter(dfBuffaloTrophy['长度(cm)'], dfBuffaloTrophy['重量(kg)'])
plt.xlabel('长度(cm)')
plt.ylabel('重量(kg)')
plt.show()
```
