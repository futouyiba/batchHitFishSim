```python
dataXlsxPath = "tool_hook_fish_20250528172514.xlsx"
sheetName = "tool_hook_fish_20250528172514"

import pandas as pd

df = pd.read_excel(dataXlsxPath, sheet_name=sheetName)

df.head()
```

```python
# 只分析 “鱼名称” 列 为 “Bigmouth_Buffalo_Trophy” 的行
# 计算 “Bigmouth_Buffalo_Trophy” 出现的次数
dfBuffaloTrophy = df[df['鱼名称'] == 'Bigmouth_Buffalo_Trophy']
countBuffaloTrophy = len(dfBuffaloTrophy)
print("Bigmouth_Buffalo_Trophy 出现的次数为：", countBuffaloTrophy)
```

```python
# 画出其中的 重量(kg) 列 和 长度(cm) 列的二维关联曲线，横轴为长度，纵轴为重量
import matplotlib.pyplot as plt
import numpy as np

# 使用dfBuffaloTrophy中的数据
# dfBuffaloTrophy = pd.read_csv('buffaloTrophy.csv')
# dfBuffaloTrophy = dfBuffaloTrophy.dropna()
# dfBuffaloTrophy = dfBuffaloTrophy.reset_index(drop=True)
# 画出其中的 重量(kg) 列 和 长度(cm) 列的二维关联曲线，横轴为长度，纵轴为重量
plt.scatter(dfBuffaloTrophy['长度(cm)'], dfBuffaloTrophy['重量(kg)'])
plt.xlabel('长度(cm)')
plt.ylabel('重量(kg)')
plt.show()
```

```python
# 只分析 “鱼名称” 列 以 “Bigmouth_Buffalo” 开始的行
# 添加na=False参数处理空值
dfBigmouthBuffaloAll = df[df["鱼名称"].str.startswith("Bigmouth_Buffalo", na=False)]

import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # Windows系统字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

plt.figure(figsize=(10, 6))
plt.scatter(dfBigmouthBuffaloAll['长度(cm)'], 
            dfBigmouthBuffaloAll['重量(kg)'],
            alpha=0.6,
            edgecolor='w')

plt.xlabel('鱼体长度(cm)')
plt.ylabel('鱼体重量(kg)')
plt.title('Bigmouth Buffalo 长度-重量分布关系')
plt.grid(True, linestyle='--', alpha=0.6)

# 保存图片到当前目录
plt.savefig('length_weight_correlation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 对于dfBigmouthBuffaloAll中的“鱼名称”列，统计每个名称出现的次数，并按照出现次数从大到小排序
import pandas as pd
dfBigmouthBuffaloAll['鱼名称'].value_counts().sort_values(ascending=False)
```

```python
from matplotlib import cm

plt.figure(figsize=(12, 7))

# 创建颜色映射
unique_fish = df['鱼名称'].unique()
colors = cm.tab10(np.linspace(0, 1, len(unique_fish)))

# 分组绘制散点图
for fish, color in zip(unique_fish, colors):
    mask = df['鱼名称'] == fish
    plt.scatter(dfBigmouthBuffaloAll.loc[mask, '长度(cm)'],
                dfBigmouthBuffaloAll.loc[mask, '重量(kg)'],
                c=[color],
                label=fish,
                alpha=0.7,
                edgecolors='w',
                s=60)

plt.xlabel('鱼体长度(cm)')
plt.ylabel('鱼体重量(kg)')
plt.title('不同鱼种长度-重量分布关系')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle=':', alpha=0.5)

# 保存高清图片
plt.savefig('multi_species_plot.png', 
            dpi=300, 
            bbox_inches='tight',
            transparent=True)
plt.show()
```

```python
# 取某个特定重量的Bigmouth_Buffalo鱼，查看其重量的最大值、最小值
sampleWeight = 1.1
dfBigmouthSampleWeight = dfBigmouthBuffaloAll[dfBigmouthBuffaloAll['重量(kg)'] == sampleWeight]
print('最大长度：', dfBigmouthSampleWeight['长度(cm)'].max())
print('最小长度：', dfBigmouthSampleWeight['长度(cm)'].min())
```

```python
tolerance = sampleWeight * 0.01  # 1%浮动范围
lower_bound = sampleWeight - tolerance
upper_bound = sampleWeight + tolerance

dfBigmouthSampleWeight = dfBigmouthBuffaloAll[
    (dfBigmouthBuffaloAll['重量(kg)'] >= lower_bound) 
    & (dfBigmouthBuffaloAll['重量(kg)'] <= upper_bound)
]

print(f'符合重量范围({lower_bound:.2f}~{upper_bound:.2f}kg)的记录数：', len(dfBigmouthSampleWeight))
print('最大长度：', dfBigmouthSampleWeight['长度(cm)'].max())
print('最小长度：', dfBigmouthSampleWeight['长度(cm)'].min())
```
